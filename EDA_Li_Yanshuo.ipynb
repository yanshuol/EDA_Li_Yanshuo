```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Read the csv file using Pandas. Store the output into a dataframe.
data = pd.read_csv('Mall_Customers.csv')
frame = pd.DataFrame(data)
print(frame)
frame.info()
```

```
# sort by gender
frame.sort_values(by='Age', ascending=True, inplace=True)
print(frame)
```

```
# create some subplots
fig, axs = plt.subplots(1,2)
```

```
# Generate a bar chart to show Gender difference
freq = frame['Genre'].value_counts().plot(kind = 'bar', figsize=(15,8), title = 'Gender Difference', color = ('orange','c'))
freq.set_xlabel('Gender')
freq.set_ylabel('Freqneucy')
# frame.set_index(['Genre', 'Age']).count(level='Genre')
```

```
# Generate a pie chart to show Gender proportion
freq = frame['Genre'].value_counts().plot.pie(figsize=(20,10),title = 'Proportion by Gender', colors = ['orange','c'])
freq.set_ylabel('')
freq.legend()
```

```
freq = frame['Age'].plot(kind = 'hist', bins=10, range=[15,70], figsize=(15,8), title = 'Histogram of Age', color = 'orange')
freq.set_xlabel('Age')
freq.set_ylabel('Freqneucy')
```

```
# Pie Chart of Age
#freq = frame['Age'].value_counts().plot(kind = 'pie', autopct='%.1f%%', startangle=45, figsize=(15,8), title = 'Pie Chart of Age')
#freq.set_xlabel('Age')
#freq.set_ylabel('Freqneucy')
```

```
freq = frame.boxplot('Age', figsize=(10,10), color = 'R')
```

```
freq = frame['Annual Income (k$)'].plot(kind = 'hist', bins=10, range=[10,140], figsize=(15,8), title = 'Histogram of Annual Income', color = 'orange')
freq.set_xlabel('Annual Income')
freq.set_ylabel('Freqneucy')
```

```
freq = frame['Spending Score (1-100)'].plot(kind = 'hist', bins=10, range=[10,110], figsize=(15,8), title = 'Histogram of Spending Score', color = 'c')
freq.set_xlabel('Spending Score')
freq.set_ylabel('Freqneucy')
```

```
freq = frame['Annual Income (k$)'].plot(kind = 'density', figsize=(15,8), title = 'Distribution of Annual Income & Spending Score', color = 'orange')
freq1 = frame['Spending Score (1-100)'].plot(kind = 'density', figsize=(15,8), color = 'C')
freq1.set_xlabel('Class')
freq1.set_ylabel('Freqneucy')
freq1.legend()
```

```
# create some subplots
freq = frame[(frame['Genre'] == 'Male')].boxplot('Annual Income (k$)', figsize=(10,10), color = 'c')
freq.set_xlabel('Annual Income of Male')
freq.set_ylabel('Freqneucy')
```

```
freq = frame[(frame['Genre'] == 'Female')].boxplot('Annual Income (k$)', figsize=(10,10), color = 'orange')
freq.set_xlabel('Annual Income of Female')
freq.set_ylabel('Freqneucy')
```

```
frame2 = frame[(frame['Genre'] == 'Female')]
freq = frame2['Annual Income (k$)'].plot(kind = 'hist', bins=10, range=[10,140], figsize=(15,8), title = 'Histogram of Female Annual Income', color = 'orange')
freq.set_xlabel('Annual Income')
freq.set_ylabel('Freqneucy')
```

```
frame['Amount of Customers'] = frame['CustomerID'].value_counts()
frame1 = frame.groupby('Genre').sum()[['Age','Annual Income (k$)','Spending Score (1-100)','Amount of Customers']]
frame1['Average Annual Income'] = frame1['Annual Income (k$)'] / frame1['Amount of Customers']
frame1['Average Spending Score'] = frame1['Spending Score (1-100)'] / frame1['Amount of Customers']
print(frame1)
```

```
freq = frame1['Average Annual Income'].plot(kind = 'bar', figsize=(4,10), title = 'Average Annual Income Difference', color = ('orange','c'))
freq.set_xlabel('Gender')
freq.set_ylabel('Average Annual Income')
```

```
freq1 = frame1['Average Spending Score'].plot(kind = 'bar', figsize=(4,10), title = 'Average Spending Score Difference', color = ('orange','c'))
freq1.set_xlabel('Gender')
freq1.set_ylabel('Average Spending Score')
```
